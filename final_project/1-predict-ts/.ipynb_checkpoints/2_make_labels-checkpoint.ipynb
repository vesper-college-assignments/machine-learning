```python
import pandas as pd
import os
```

```python
data = pd.read_csv('../data/CMAPSSData/data_all.csv', parse_dates=['time'])
data.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# criar os labels a partir de RUL
data['target'] = data['RUL'].apply(lambda x: 1 if x <=40 else 0)
data = data.drop(['RUL'], axis=1)
data.head(5)
```

```python
# Generate X and y
y = data['target']
X = data.drop(['target'], axis=1)
```

```python
# Persist
os.remove('data.h5')
X.to_hdf('X.h5', key='X', mode='w')
y.to_hdf('y.h5', key='y', mode='w')
```
